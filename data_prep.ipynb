```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Lets take a look at the Values for each coloumn.
for x in mi_data.columns:
    print('Unique values in {}: '.format(x), mi_data[x].unique())
    print("\n\n")
```

Lets remove the columns that are infeasible for plotting.
I will also remove a few columns that are not suited for the current usecase.

```python
columns_to_drop = ['Role', 'Country of Residence', 'ResponseID', 'Mental Illness (Precieved)', 'Genetic Mental Illness']

for column in columns_to_drop:
    if column in mi_data.columns:
        mi_data = mi_data.drop(columns=[column])
    else:
        print("{} is not in the Dataset!".format(column))
```

```python
mi_data.columns
```

Since I am trying to visualize data that signifies the presence of mental illneses in a Tech company 
I will only select the rows that are from a tech company using the 'Tech Company' column

```python
mi_data.drop(mi_data[mi_data['Tech Company'] == False].index, inplace=True)

# Since all the remaining values are True in Tech Company, it is not required.

mi_data.drop(columns=['Tech Company'], inplace=True)

mi_data.columns
```

Lets remove any remaining odd values.


```python
mi_data.drop(mi_data[mi_data['Age'] == 'nan'].index, inplace=True)
```

```python
columns = pd.concat([mi_data.isnull().sum(), mi_data.nunique(), mi_data.dtypes], axis=1, sort=False)
columns
```

Since the number of missing values is not much we can drop these values.

```python
mi_data.drop(mi_data[mi_data['Age'].isnull()].index, inplace=True)
```
